## Le Mans Data Analysis

```python
import pandas as pd
import numpy as np
import datetime
```

```python

DataFolder = "./DATA/"
lmDat2017Race = DataFolder + "2017-LeMans-23_Analysis_Race.csv"

# data file is not actually comma sep -- it's actually semi-colons

dfLM2017Race = pd.read_csv(lmDat2017Race, sep = ";")
```

```python
dfLM2017Race
```

```python
# lapTimeMS and other time columns are "strings" with M:ss.sss (some have no M:)
dfLM2017Race.columns=['carNum', 'driverNum', 'lapNum', 'lapTimeMS',
                     'lapImp', 'crossFinishLineInPit', 's1',
                     's1Imp', 's2', 's2Imp', 's3', 's3Imp',
                     'kph', 'elapsed', 'hour', 's1Large', 's2Large', 's3Large',
                     'topSpeed', 'driverName', 'pitTime', 'class', 'group', 'team',
                     'manufacturer', 'other']
dfLM2017Race
```

```python
lapTimeMS = dfLM2017Race['lapTimeMS']
# need to convert s1Str -> s1Float

lapTime = lapTimeMS.apply(lambda x: datetime.datetime.strptime(x,'%H:%M:%S.%f'))


lapTime
```
